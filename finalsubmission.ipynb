```python
childmortality=pd.read_csv("/content/child_mortality_0_5_year_olds_dying_per_1000_born (1).csv")
income=pd.read_csv("/content/income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
```

```python
childmortality.head()
```

```python
income.head()
```

```python
number_per_woman=df[ ["country","2023"] ]

childmortality1=childmortality[ ["country","2023"] ]

income1=income[ ["country","2023"] ]
```

```python
all_data_final1 =pd.merge( number_per_woman,childmortality1,"left", on=["country"])
all=pd.merge(all_data_final1,income1,"left", on=["country"])
all1 = all.rename(columns={'country':"Country", "2023_x":'ChildPerWoman',"2023_y": "ChildMortality", "2023": 'Income'})

x=all1.fillna(0)
x.head(2)
```

```python
x["Income"]=x["Income"].replace({"k":"*1e3"}, regex=True).map(pd.eval).astype(float)
```

```python
result=x
result.head(3)
```

```python
geo=pd.read_csv("geo.csv")
geo = geo.rename(columns={'name':"Country"})
geo.head(1)
```

```python
final_result =pd.merge( result,geo,"left", on=["Country"])
final_result1=final_result
final_result1.head(2)
```

```python
final_result2 = final_result1[["Country","ChildPerWoman","ChildMortality","Income","four_regions"]]
final_result2 = final_result2.rename(columns={"four_regions": "Major Region"})
final_result2.head(2)
```

```python
from IPython.core.pylabtools import figsize
from matplotlib import colors
import seaborn as sns
plt.rcParams['figure.figsize']=(10,5)
# use the scatterplot function to build the bubble map
sns.scatterplot(data=final_result2, x="Income", y="ChildPerWoman", size="ChildMortality", sizes=(20, 2000),hue="Major Region",alpha=0.4).set(title="Regions are still Facing High Child Mortality Rates " )
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0, handlelength=8, handleheight=4, fontsize = 10);
# show the graph
plt.grid();
plt.show();
```

Let's embark on a historical journey from 1799 to the present, witnessing humanity's remarkable progress. However, even amidst this astonishing development, certain regions persist in grappling with the burden of high mortality and fertility rates, reminding us of the challenges that endure.

```python
print(income.columns)
print(childmortality.columns)
print(df.columns)
```

```python
income_long = income.melt(id_vars=["country"], var_name="year", value_name="income")
childmortality_long = childmortality.melt(id_vars=["country"], var_name="year", value_name="childmortality")
fertility_long = df.melt(id_vars=["country"], var_name="year", value_name="fertility")

df_long = income_long.merge(childmortality_long, on=["country", "year"]).merge(fertility_long, on=["country", "year"])
df_long1 = df_long.rename(columns={'country':"Country"})

df_long1
```

```python
Combine = df_long1.merge(geo, on="Country")
Combine.head(2)
```

```python
Combine1 = Combine[["Country","year","fertility","childmortality","income","four_regions"]]
#Giving the desired name to one of the column
Combine1 = Combine1.rename(columns={"four_regions": "Major Region"})
# Convert "year" column to integer
Combine1["year"] = Combine1["year"].astype(int)
# Filter data after 2023
Combine1 = Combine1[Combine1["year"] <= 2023]
Combine1.head()
```

```python
Combine1.shape
```

```python
import plotly.express as px
```

```python
Combine1 = Combine1.dropna(subset=['childmortality'])

fig = px.scatter(Combine1,x="income", y="fertility",animation_frame="year",
 animation_group="Country",size="childmortality",
 color="Major Region",
  hover_name="Country",log_x=True,
 size_max=25,range_x=[200,150000], range_y=[0,10]
 )
# Tune marker appearance and layout
fig.update_traces(mode="markers", marker=dict(sizemode="area",
 ))
fig.update_layout(
 title="Despite Progress:Some regions are still facing high Child Mortality Rates",
 xaxis=dict(
 title='Income(Adjusted GDP per Capita)',
 gridcolor='white',
 gridwidth=2,
 type='log',
 ),
 yaxis=dict(
 title='Number of children per women',
 gridcolor='white',
 gridwidth=2,
 ),
 paper_bgcolor='rgb(243, 243, 243)',
 plot_bgcolor='rgb(243, 243, 243)',
)
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 700
fig.show()
```
